```python
import pandas as pd
import re
import matplotlib.pyplot as plt
import seaborn as sns
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
import string
import itertools
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import warnings
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, confusion_matrix, f1_score, recall_score
from sklearn.linear_model import PassiveAggressiveClassifier

warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("news.csv")
```

```python
df.head()
```

```python
def preprocess(text):
    text = ''.join(''.join(s)[:2] for _, s in itertools.groupby(text))
    text = re.sub(r'[^A-Za-z\s]',r'',text)
    text = re.sub(r'\n',r'',text)
    tokens = nltk.word_tokenize(text)
    tokens = [word.lower() for word in tokens]
    stop_words = stopwords.words('english')
    token_list = []
    for word in tokens:
        if not word in stop_words:
            token_list.append(word)
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(token, pos="v") for token in tokens]
    s = ' '.join([str(elem) for elem in tokens]) 
    return s

def labels(value):
    if value == 'FAKE':
        l = 1
    elif value == 'REAL':
        l = 0
    return l
```

```python
df['clean_label'] = df['label'].apply(lambda x: labels(x))
df['clean_text'] = df['text'].apply(lambda x: preprocess(x))
```

```python
df = df.drop(['Unnamed: 0', 'text', 'title', 'label'], axis=1)
```

```python
X = df.clean_text
y = df.clean_label
```

```python
train_x, test_x, train_y, test_y = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
l = stopwords.words('english')
```

```python
cv = CountVectorizer(stop_words=l)
X_train_cv = cv.fit_transform(train_x)
X_test_cv = cv.transform(test_x)
```

```python
tf = TfidfVectorizer(stop_words=l, max_df = 0.8, min_df = 0.01)
X_train_tf = tf.fit_transform(train_x)
X_test_tf = tf.transform(test_x)
```

```python
tf.get_feature_names()
```

```python
naive_bayes_cv = MultinomialNB()
naive_bayes_cv.fit(X_train_cv, train_y)
predictions = naive_bayes_cv.predict(X_test_cv)
print('Accuracy score: ', accuracy_score(test_y, predictions))
print('Precision score: ', precision_score(test_y, predictions, average='weighted'))
print('Recall score: ', recall_score(test_y, predictions, average='weighted'))
print('F1 score: ', f1_score(test_y, predictions, average='weighted'))
```

```python
cm_cv = confusion_matrix(test_y, predictions)
print(cm_cv)
fig, ax = plt.subplots(figsize=(5, 5))
ax.set(yticks=[-0.5, 1.5], 
       xticks=[0, 1])
sns.heatmap(cm_cv, square=True, annot=True, cmap='YlGnBu', cbar=True, ax = ax)
```

```python
naive_bayes_tf = MultinomialNB()
naive_bayes_tf.fit(X_train_tf, train_y)
predictions = naive_bayes_tf.predict(X_test_tf)
print('Accuracy score: ', accuracy_score(test_y, predictions))
print('Precision score: ', precision_score(test_y, predictions, average='weighted'))
print('Recall score: ', recall_score(test_y, predictions, average='weighted'))
print('F1 score: ', f1_score(test_y, predictions, average='weighted'))
cm = confusion_matrix(test_y, predictions)
print(cm)
sns.heatmap(cm, annot=True, cmap='YlGnBu', cbar=True, xticklabels=['FAKE', 'REAL'], yticklabels=['FAKE', 'REAL'])
```

```python
#count vectorizer clearly has a better accuracy as compared to tf idf
```

```python
pac = PassiveAggressiveClassifier(max_iter=50)
pac.fit(X_train_cv, train_y)
predictions = pac.predict(X_test_cv)
print('Accuracy score: ', accuracy_score(test_y, predictions))
print('Precision score: ', precision_score(test_y, predictions, average='weighted'))
print('Recall score: ', recall_score(test_y, predictions, average='weighted'))
print('F1 score: ', f1_score(test_y, predictions, average='weighted'))
cm = confusion_matrix(test_y, predictions)
print(cm)
```
